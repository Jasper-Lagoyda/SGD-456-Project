```python
import numpy as np
import matplotlib.pyplot as plt

# Rosenbrock function
def rosenbrock(x, y):
    return (1 - x)**2 + 100 * (y - x**2)**2

# Gradient of Rosenbrock function
def rosenbrock_gradient(x, y):
    dx = -2 * (1 - x) - 400 * x * (y - x**2)
    dy = 200 * (y - x**2)
    return np.array([dx, dy])

# Batch Gradient Descent
def batch_gradient_descent(gradient_fn, lr, num_iterations):
    x = np.array([-2, 2])  # Initial point
    trajectory = [x]
    for i in range(num_iterations):
        gradient = gradient_fn(x[0], x[1])
        x = x - lr * gradient
        trajectory.append(x)
    return np.array(trajectory)

# Stochastic Gradient Descent
def stochastic_gradient_descent(gradient_fn, lr, num_iterations):
    x = np.array([-2, 2])  # Initial point
    trajectory = [x]
    for i in range(num_iterations):
        random_gradient = gradient_fn(*x* np.random.random(2))
        x = x - lr * random_gradient
        trajectory.append(x)
    return np.array(trajectory)

# Common parameters
learning_rate = 0.001
max_iterations = 10000

# Generate contour plot
x = np.linspace(-2.0, 2.0, 100)
y = np.linspace(-1.0, 3.0, 100)
X, Y = np.meshgrid(x, y)
Z = rosenbrock(X, Y)

# Calculate minimum point of the Rosenbrock function
min_point = np.array([1, 1])  # Minimum point of the function

# Batch Gradient Descent
batch_traj = batch_gradient_descent(rosenbrock_gradient, lr=learning_rate, num_iterations=max_iterations)

# Stochastic Gradient Descent
stochastic_traj = stochastic_gradient_descent(rosenbrock_gradient, lr=learning_rate, num_iterations=max_iterations)

plt.figure(figsize=(8, 6))

plt.subplot(1, 2, 1)
plt.contour(X, Y, Z, levels=np.logspace(-1, 3, 20))
batch_x_vals = batch_traj[:, 0]
batch_y_vals = batch_traj[:, 1]
plt.scatter(batch_x_vals, batch_y_vals, c='red', label='Batch GD')
plt.plot(batch_x_vals, batch_y_vals, linestyle='-', color='red')  # Add lines
plt.scatter(min_point[0], min_point[1], c='green', label='Minimum', marker='o')  # Add minimum point
plt.title('Batch Gradient Descent on Rosenbrock Function')
plt.xlabel('X')
plt.ylabel('Y')
plt.legend()
plt.xlim(-2, 2)  # Adjust the X-axis limits
plt.ylim(-1, 3)  # Adjust the Y-axis limits

plt.subplot(1, 2, 2)
plt.contour(X, Y, Z, levels=np.logspace(-1, 3, 20))
stochastic_x_vals = stochastic_traj[:, 0]
stochastic_y_vals = stochastic_traj[:, 1]
plt.scatter(stochastic_x_vals, stochastic_y_vals, c='blue', label='SGD')
plt.plot(stochastic_x_vals, stochastic_y_vals, linestyle='-', color='blue')  # Add lines
plt.scatter(min_point[0], min_point[1], c='green', label='Minimum', marker='o')  # Add minimum point
plt.title('Stochastic Gradient Descent on Rosenbrock Function')
plt.xlabel('X')
plt.ylabel('Y')
plt.legend()
plt.xlim(-2, 2)  # Adjust the X-axis limits
plt.ylim(-1, 3)  # Adjust the Y-axis limits

plt.tight_layout()
plt.show()
```
